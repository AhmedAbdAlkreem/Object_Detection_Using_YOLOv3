# Imports

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cv2.__version__
```

# Helper functions

```python
# Function to read image.
def imread(path):
    return cv2.imread(path)

# To save the image in specific path.
def imwrite(img, path):
    if cv2.imwrite(path, img):
        print('Image saved to "{}"'.format(path))
    else:
        print('Failed to save image to "{}"'.format(path))

# To show the image.
def imshow(img, figsize=(10, 10), is_bgr=True):
    if is_bgr:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.imshow(img, cmap='gray')
    plt.show()
```

# YOLO

```python
# Call the class from file my_yolo.py
from my_yolo import MyYolo
```

```python
yolo = MyYolo('model/')
```

## Predict from image

```python
# Read the image.
img = imread('test.jpg')

# Show the result of image.
imshow(img, figsize = (7, 7))
```

```python
# Predict the result of prediction using yolov3.
predictions, result = yolo.predict(img)

# Show the result.
imshow(result)
```

```python
# Save the result of image.
imwrite(result, 'result.jpg')
```
